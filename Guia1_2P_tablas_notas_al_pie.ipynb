# Notas al pie
## Concepto rápido
> Es una nota aclaratoria que se ubica al final de la misma pagina donde se encuentra el parrafo citado

## Ubicación
> Se encuentra dentro del __menu ```referencias```__.

## Aplicación
> - Ubicarse en la ultima palabra del texto a referenciar
> - Clic en ```referencias```, y despues en  ```insertar nota al pie```.
> - Word dirige automaticamente al final de la pagina, en esa ubicacion se debe escribir el texto complementario

# Notas al final
## Concepto rápido
> Tiene el mismo uso de las notal al pie, difieren en la ubicacion, este tipo de notas seran mostradas al final del documento
