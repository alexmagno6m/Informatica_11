# Tablas indexadas
## Conceptos Basicos
Son aquellas agrupaciones de elementos como: figuras, graficos, ecuaciones, ilustraciones y las propias tablas que aparecen a lo largo de un documento profesional, academico o de caracter investigativo

> ### Tablas de ilustraciones
>
> Una ilustracion es producto de un diseÃ±o artifical, el cual pudo ser eleborado en forma manual, herramientas digitales, o una combinacion de ambos. Dentro de esta categoria entran imagenes de comics, cuadros artisticos, pinturas, etc.
>
> ### Tablas de imagenes
> 
> Corresponde a las fotografias propiamente dichas del __mundo real__, manipuladas o no, que son captadas por lentes de camaras de aparatos aptos para esa labor.
>
> ### Tablas de figuras
>
> Son diagramas elaborados con el fin de complementar la informacion al texto, contribuir a su explicacion, aclarar conceptos o modelar sistemas, entran en esta categoria los diagramas de flujo, mapas mentales, diagramas de nodos, graficos de funciones matematicas o cientificas.
>
> ### Tablas de ecuaciones
>
> Toda expresion matematica o cientifica como una ecuacion, funcion, sumatoria o la notacion numerica y simbolica especial debe ir agrupada en esta categoria.

## Procedimiento para referenciar tablas indexadas
Como se ha visto hasta el momento, para indexar una tabla primero debemos referenciarla con el correspondiente titulo, es conocido en Word como rotulo o etiqueta.
> ### Insertar el titulo del elemento a indexar en la tabla
> - Se hace en el menu referencias, seleccionado _insertar titulo_.
> - Si el titulo no se ha creado, se debe proceder a agregar el correspondiente a la categoria que se necesite.
> - Se debe cambiar el estilo para que corresponda con el resto del documento.
>
> ### Generar la tabla indexada
> Las tablas indexadas deben ubicarse inmediatamente despues de la tabla de contenido. Una vez alli siga los siguientes pasos:
> - Haga clic en referencias, _insertar tabla de ilustraciones_
> - Seleccione la etiqueta o rutulo de la categoria de la tabla a indexar
> - Haga clic en aceptar
> - Corrija el estilo para que se ajuste al resto del documento.

# Referencias Cruzadas

## Concepto
Son enlaces de elementos referenciados en el documento que se citan en un lugar usualmente distitno a donde esta ubicado tal elemento.

## Procedimiento para crear referencias cruzadas
Lo primero es haber creado previamente el elemento a referenciar, entiendase: titulos, tablas de contenido, elementos indexados.
> - Se selecciona el texto correspondiente, en general debe mencionar explicitamente el elemento a referenciar, por ejemplo: _vease tabla 1_.
> - En el menu referencias, se selecciona _referencia cruzada_.
> - En la ventana se escoge la categoria del elemento a referenciar, por ejemplo: titulo, ilustraciones, imagenes, etc.
> - En la lista se selecciona el elemento en especifico.
