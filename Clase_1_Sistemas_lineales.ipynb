# SOLUClON GRAFICA SISTEMAS LINEALES
## Conceptos claves
Un sistema de ecuaciones lineales corresponde a un conjunto de funciones de este tipo relacionadas con un punto en comun llamado __solucion__.
> ### Solucion Grafica
Corresponde al punto de interseccion de todas las rectas (o funciones) que componen el sistema:

> _Fuente_: [Google](https://a9dc6584-a-62cb3a1a-s-sites.googlegroups.com/site/matematicasjuanmanuelista/matematicas-9/algebra-9/2-sistemas-de-ecuaciones-lineales/1-solucion-grafica-de-un-sistema-de-ecuaciones-lineales-2x2/Imagen40.png?attachauth=ANoY7cpM22xZIyv4zMHncQYjGxztsM9i7dxuH9vl2PVIoVCOGE5G26f-lzzFdwBxGJSRshOP1NYTIER-9swn7NC4hmHcmpdImcT7N9kVBA-5hyDKPkENagECmuW6A_tlW_v_JgDq4VGUqSHJaO2hkQXQ-ZlIhtN2ghJpQB5uYD5Tbp-Y_bVB9C_iICV0JtfJgQNTWafK720pUaSTCVsExxNebg4HB7MVRctNCC6SlbbNNaW47_lO1CdMKdx-UUwitqB_xjwmqmzGt00586jRmI61KfZtAN8b1aOdG0nFaEY5ORAtd06hOgekzLZe3IhyKeQoStkssh3pa7ESIk08bLmu0MFli_5zw51MQr3Soh11-GgnJfk-muHv-MdA3M49Dk15jtVaQU6WVmPkChSDEjLN5VWEnP-Y4w%3D%3D&attredirects=0)<br>
>
> Cuando las graficas de las funciones en un sistema lineal no se interceptan se dice que __no tienen solucion__
>
> ### Punto de interseccion
> Es el espacio grafico donde se obseva que las rectas se cruzan
> ### Sistemas Compatibles
> Cuando un sistemas tiene una y solo una solucion
>
> ### Sistemas Incompatibles
>
> Si el sistema tiene infinitas soluciones, como el caso de ecuaciones equivalentes, se dice que es un sistema incompatible indeterminado.
>
> Si el sistema NO tiene soluciones, en la mayoria de casos por ser funciones paralelas una respecto a la otra. Se le conoce como sistema incompatible.

## Ejemplo a desarrollar
### Objetivo
Hallar la solucion grafica (si es posible) de un sistema lineal de dos ecuaciones, usando Microsoft Excel.

### Sistema a solucionar
$$ \left\{\begin{matrix}
y=3x+2 & \\ 
y=6-5x 
\end{matrix}\right.$$

### Primera etapa: preparando los datos numericos
1. En Excel halle los valores de X e Y segun lo explicado por el profesor en la clase anterior:
$$ y = 3x + 2 $$

> - Fije Valor Inicial en 0 __Celda E1__
>
> - Incremento en 1 __Celda E2__
>
> - Arrastre hasta la __fila 22__
>
> - Si calculo bien los valores, su hoja de excel debe verse de la siguiente forma:
>> 

2. En la columna C halle los valores para la siguiente ecuacion, use los mismos datos de X (columna A) que empleo en la primera funcion
$$ y = 6 - 5x $$

### Segunda etapa: graficando el sistema
1. Seleccione las columnas con los valores (A, B y C)
2. Inserte el grafico de dispersion indicado en la clase anterior

>Su grafica debe verse asi
>

3. Modifique el titulo del grafico y escriba las ecuaciones de las dos graficas
4. Conteste la siguiente pregunta en su cuaderno:

> - Segun lo leido al inicio, para este sistema existe una solucion grafica?. (Justifique su respuesta)
> - Es posible determinar con __visualmente y con exactitud__ el punto (valor de X) donde ambas graficas se interceptan?

### Tercera etapa: determinando la solucion del sistema
Al explorar visualmente el sistema, la solucion se encuentra cerca a los valores de X entre 0 y 3. Este segmento lo llamaremos __zona de interes__ y modificando los valores de __valor inicial__ e __incremento__ procederemos a ampliar dicha zona.
1. Cambie el valor inicial a -1
2. Cambie el incremento a 0.5
3. Observe la grafica, seleccione la funcion de la linea azul, luego ubique el puntero del mouse sobre el punto donde las graficas se cruzan, mire las coordenadas que se muestran.
4. Ahora selecciona la linea naranja, ubique el puntero en el mismo sitio, __si las coordenadas son las mismas__, se ha encontrado la solucion al sistema.
5. Conteste en su cuaderno la siguiente pregunta:

> - Encontro la solucion al sistema?, si es asi, escriba las coordenadas, es decir los valores de X e Y

__Nota__: Cuando la solucion aun no es clara, se deben seguir modificando el __valor inicial__ teniendo en cuenta que se ubique lo mas cercano y a la izquierda del punto de interseccion.

El __incremento__ siempre debe ir disminuyendo hasta que se encuentre la solucion, los valores recomendados para cambiar este valor son:

| Cambios Incremento  | Valor  |
|---|---|
|  Primer cambio |  0.5 |
| Segundo cambio  |  0.2 |
|  Tercer cambio |  0.1 |
| Cuarto cambio  |  0.05 |
|  Quinto cambio |  0.01 |

## Formulas y proceso usados en clase
> La columna de valores X debe contener ahora un formato condicional que aproxime el resultado a una diferencia $\approx 0.04$, para lograrlo en la primera celda (usualmente A2) se ingresa a formato condicional:
>
> - Formato condicional > Nueva regla > Usar formula que determine el formato para usar en las celdas

> $$ =ABS(B2-C2)<=0.004 $$
